```python
import numpy as np
import librosa as lr
import matplotlib.pyplot as plt
import os
```

```python
path = "/home/pjja2556/june/ai_project/NIKL_DIALOGUE_2020_PCM_v1.3_part1/pcm/SDRW2000000001"
```

```python
pcm_list = os.listdir(path)
```

```python
len(pcm_list)
```

```python
### Load File
target = os.path.join(path, pcm_list[0])
f=open(target, 'rb')
buf = bytearray(f.read())
pcm_data = np.frombuffer(buf, dtype = 'int16')			# bit 수에 맞게 입력
wav_data = lr.util.buf_to_float(x=pcm_data, n_bytes=2)

### Draw Graph
x_axis = [i/8000 if i>0 else i for i in range(0, len(wav_data))]	# sampling rate에 맞게 x축 시간값 저장
plt.figure()
plt.plot(x_axis, wav_data)		# x축은 x_axis, y축은 wav_data 사용
plt.show()
```
