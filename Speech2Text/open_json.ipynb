```python
import os
import json
import pandas as pd
```

```python
# JSON 파일이 있는 폴더 경로
folder_path = "../NIKL_DIALOGUE_2020_PCM_v1.3_part4/json"

# 파일 목록 가져오기
file_list = os.listdir(folder_path)
json_files = [f for f in file_list if f.endswith(".json")]
```

```python
json_files
```

```python
speaker_dict = {}
utterance_list = []

for json_file in os.listdir(folder_path):
    if json_file.endswith('.json'):
        with open(os.path.join(folder_path, json_file), 'r', encoding='utf-8') as f:
            data = json.load(f)
        
        for speaker in data['document'][0]['metadata']['speaker']:
            speaker_dict[speaker['id']] = {
                'age': speaker['age'],
                'sex': speaker['sex']
            }
        
        for d in data['document'][0]['utterance']:
            utterance_list.append({
                'id': d['id'],
                'form': d['form'],
                'original_form': d['original_form'],
                'speaker_id': d['speaker_id'],
                'start': d['start'],
                'end': d['end'],
                'note': d['note']
            })

for i, row in enumerate(utterance_list):
    speaker_id = row['speaker_id']
    utterance_list[i]['age'] = speaker_dict[speaker_id]['age']
    utterance_list[i]['sex'] = speaker_dict[speaker_id]['sex']

df = pd.DataFrame(utterance_list)
```

```python
df = df.drop(['note'],axis=1)
```
