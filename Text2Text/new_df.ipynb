```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("./input_df.csv")
df
```

```python
for i in range(len(df)):
    if '?' in df['value'][i]:
        print("value : " + df['value'][i])
        print("label : " + df['label'][i])
        
```

```python
# '?'가 있는 행만 선택
df_selected = df[df['value'].str.contains('\?', na=False)]

# 새로운 데이터프레임 생성
new_df = pd.DataFrame({
    'value' : df_selected['value'].str.replace('?', '', regex=False),
    'label': df_selected['label'].str.replace('.', '', regex=False)
})

# 인덱스를 재설정합니다.
new_df = new_df.reset_index(drop=True)
new_df
```

```python
import matplotlib.pyplot as plt

# 데이터를 정의합니다.
labels = ['Remaining', 'Removed']
sizes = [len(new_df), len(df) - len(new_df)]  # Remaining과 Removed의 개수
total = sum(sizes)
colors = ['lightcoral', 'skyblue']  # 각 파이 슬라이스의 색깔
explode = (0.1, 0)  # 첫 번째 슬라이스를 조금 돌출시킵니다.

# 파이 차트를 그립니다.
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
        autopct=lambda p: '{:.1f}%  ({:.0f})'.format(p,(p/100)*total),
        shadow=True, startangle=140)

# 차트를 원형으로 유지하도록 합니다.
plt.axis('equal')

# 제목을 추가합니다
plt.title(f'Total: {total}')

# 차트를 보여줍니다.
plt.show()
```

```python
new_df.to_csv('./new_df.csv', index=False)
```
