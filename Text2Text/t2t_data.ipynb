```python
import os
import json
import pandas as pd
from konlpy.tag import Okt
```

```python
def read_json_files(folder_path):
    file_list = os.listdir(folder_path)  # 폴더 내 파일 목록 가져오기
    json_files = [file for file in file_list if file.endswith('.json')]  # JSON 파일 목록 필터링

    data = []  # 읽어온 JSON 데이터를 저장할 리스트
    count = 0
    for json_file in json_files:
        count += 1
        file_path = os.path.join(folder_path, json_file)
        with open(file_path, 'r') as file:
            json_data = json.load(file)  # JSON 파일 읽기
            data.append(json_data)  # 데이터 추가
        if (count == 30): break

    return data

folder_path = '../NIKL_DIALOGUE_2020_v1.3'
json_data = read_json_files(folder_path)
len(json_data)
```

```python
def get_sentence(json_data):
    sentence = []
    for i in range(len(json_data)):
        path = json_data[i]['document'][0]['utterance']
        
        sentence.append([])
        temp = ''
    
        for j in range(len(path)):
            temp += ' '
            for k in range(len(path[j]['form'])):
                temp += path[j]['form'][k]
                if (path[j]['form'][k] == '.' or path[j]['form'][k] == '?' or path[j]['form'][k] == '!'):
                    sentence[i].append(temp.strip())
                    temp = ''
    return (sentence)
    
get_sentence(json_data)
```

```python
sentence = get_sentence(json_data)
df_list = []

def make_dataframe(sentence):
    for i in range(len(sentence)):
        for j in range(len(sentence[i]) - 1):
            df_list.append({
                'value' : sentence[i][j],
                'label' : sentence[i][j + 1]
            })
        df = pd.DataFrame(df_list)
    
    return (df)

df = make_dataframe(sentence)
df
```

```python
okt = Okt()
df2 = {}

df2['token_value'] = df['value'].apply(okt.morphs)
df2['token_label'] = df['label'].apply(okt.morphs)
df2 = pd.DataFrame(df2)
df2
```
